```python
import pickle
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
step = 60
```

```python
def load_q_table(filename):
    if os.path.exists(filename):
        print("loaded the q table")
        with open(filename, 'rb') as f:
            return pickle.load(f)
```

```python
q_table_data = load_q_table("q_table.pkl")
# print(q_table)
```

```python
# Ensure the q_table_data is not None
if q_table_data is not None:
    q_table, rewards_per_episode, epsilon = q_table_data
    # Convert to DataFrame
    df = pd.DataFrame.from_dict(q_table, orient='index')

    # Reset index to flatten the DataFrame
    df.reset_index(inplace=True)

    # Rename columns for clarity
    df.columns = ['parity', 'mim', 'mip', 'disease'] + list(df.columns[4:])

    # Display the DataFrame
    display(df)
else:
    print("Q-table is empty or not loaded.")
```

```python
df_grouped = df.groupby('disease').mean()[['Keep','Cull']]
df_grouped
```

```python
# value of a replacement heifer (keeping) is
display(df[df['parity'] == 0])
replacement_value = df[df['parity'] == 0]['Keep'][0]
print(replacement_value) # <- this value is the expected max outcome of having this animal
print(replacement_value/60) # <- to month
```

```python
# parity = 1
# Open cows
df_parity_1_open = df[(df['parity'] == 1) & (df['mip'] == 0)]  # Open cows
df_parity_1_open_grouped = df_parity_1_open.groupby('mim').mean()[['Keep','Cull']].reset_index()
display(df_parity_1_open_grouped)

# Preg cows
df_parity_1_preg = df[(df['parity'] == 1) & (df['mip'] != 0)]  # preg
df_parity_1_preg_grouped = df_parity_1_preg.groupby('mim').mean()[['Keep','Cull']].reset_index()
# display(df_parity_1_preg_grouped)


# Plot the Q-values of 'Keep' and 'Cull' against 'mim'
plt.figure(figsize=(10, 5))
plt.plot(df_parity_1_open_grouped['mim'], df_parity_1_open_grouped['Keep']/60, label='Open cow Keep', marker = 'o')
plt.plot(df_parity_1_preg_grouped['mim'], df_parity_1_preg_grouped['Keep']/60, label='Preg cow Keep', marker = 'v')
plt.plot(df_parity_1_open_grouped['mim'], df_parity_1_open_grouped['Cull']/60, label='Cull', marker = 'o', linestyle='dashed') # because culling price is not related to preg or not
# plt.plot(df_parity_1_preg_grouped['mim'], df_parity_1_preg_grouped['Cull']/60, label='Preg cow Cull', marker = 'v', linestyle='dashed')

# plt.axhline(y=566.9143264239445, color='black', linestyle='--', label='replacement')

# parity = 2
# Open cows
df_parity_2_open = df[(df['parity'] == 2) & (df['mip'] == 0)]  # Open cows
df_parity_2_open_grouped = df_parity_2_open.groupby('mim').mean()[['Keep','Cull']].reset_index()
display(df_parity_2_open_grouped)

# Preg cows
df_parity_2_preg = df[(df['parity'] == 2) & (df['mip'] != 0)]  # preg
df_parity_2_preg_grouped = df_parity_2_preg.groupby('mim').mean()[['Keep','Cull']].reset_index()
# display(df_parity_2_preg_grouped)

plt.plot(df_parity_2_open_grouped['mim'], df_parity_2_open_grouped['Keep']/60, label='Open cow Keep', marker = '.')
plt.plot(df_parity_2_preg_grouped['mim'], df_parity_2_preg_grouped['Keep']/60, label='Preg cow Keep', marker = '.')
plt.plot(df_parity_2_open_grouped['mim'], df_parity_2_open_grouped['Cull']/60, label='Cull', marker = '.', linestyle='dashed') # because culling price is not related to preg or not


# Add labels and title
plt.xlabel('Month in Milk')
plt.ylabel('Q-value')
plt.title('Q-values of Keep and Cull for Parity = 1 Cows')
plt.legend()
# Set x-axis to use integer ticks
plt.xticks(df['mim'].unique())

# Display the plot
plt.show()
```
