```python
def load_pickle(filename):
    if os.path.exists(filename):
        print("loaded the q table")
        with open(filename, 'rb') as f:
            return pickle.load(f)
```

```python
def load_q_table(filename):
    # Ensure the q_table_data is not None
    q_table_data = load_pickle(filename)
    # print(q_table)if q_table_data is not None:
    q_table, rewards_per_episode, epsilon = q_table_data
    print("len (rewards_per_episode):", len(rewards_per_episode))
    print("len (epsilon):", epsilon)
    # Convert to DataFrame
    df = pd.DataFrame.from_dict(q_table, orient='index')

    # Reset index to flatten the DataFrame
    df.reset_index(inplace=True)

    # Rename columns for clarity
    df.columns = ['parity', 'mim', 'mip', 'disease'] + list(df.columns[4:])

    # Display the DataFrame
    return df

def load_q_table_no_sick(filename):
    # Ensure the q_table_data is not None
    q_table_data = load_pickle(filename)
    # print(q_table)if q_table_data is not None:
    q_table, rewards_per_episode, epsilon = q_table_data
    print("len (rewards_per_episode):", len(rewards_per_episode))
    print("len (epsilon):", epsilon)
    
    # Convert to DataFrame
    df = pd.DataFrame.from_dict(q_table, orient='index')

    # Reset index to flatten the DataFrame
    df.reset_index(inplace=True)

    # Rename columns for clarity
    df.columns = ['parity', 'mim', 'mip'] + list(df.columns[3:])

    # Display the DataFrame
    return df
```

```python
### track a cow
def track_a_cow_from_preg(df, parity, starting_mim, starting_mip):
    empty_df = pd.DataFrame(columns=['parity', 'mim', 'mip'])
    mim = starting_mim 
    mip = starting_mip
    while mip < 8:
        selected_row = df[(df['parity'] == parity) & (df['mim'] == mim) & (df['mip'] == mip)]
        empty_df = empty_df.append(selected_row, ignore_index=True)
        mim += 1
        mip += 1
    while mip == 8 or mip == 9: # dry
        selected_row = df[(df['parity'] == parity) & (df['mim'] == 0) & (df['mip'] == mip)]
        empty_df = empty_df.append(selected_row, ignore_index=True)
        mip += 1
    
#     print(len(empty_df))
    empty_df['Month after calving'] = range(starting_mim, starting_mim+9)
    
    return empty_df

def track_a_cow_open(df, parity):
    empty_df = pd.DataFrame(columns=['parity', 'mim', 'mip'])
    mim = 1
    while mim < 20:
        selected_row = df[(df['parity'] == parity) & (df['mim'] == mim) & (df['mip'] == 0)]
        empty_df = empty_df.append(selected_row, ignore_index=True)
        mim += 1
    
    empty_df['Month after calving'] = range(1, 20)
    
    return empty_df
```

```python
cow141 = track_a_cow_from_preg(df, 1, 4, 1)
display(cow141)
cow151 = track_a_cow_from_preg(df, 1, 5, 1)
cow161 = track_a_cow_from_preg(df, 1, 6, 1)
cow171 = track_a_cow_from_preg(df, 1, 7, 1)
cow181 = track_a_cow_from_preg(df, 1, 8, 1)
cow191 = track_a_cow_from_preg(df, 1, 9, 1)
cow1101 = track_a_cow_from_preg(df, 1, 10, 1)
display(cow1101)
open1 = track_a_cow_open(df, 1)
# open1

plt.figure(figsize=(10, 5))
plt.plot(cow141['Month after calving'], cow141['Diff'], label='at MIM = 4', color = 'b', marker = 'o')
# plt.plot(cow151['Month after calving'], cow151['Diff'], label='151 cow', color = 'r', marker = 'o')
# plt.plot(cow161['Month after calving'], cow161['Diff'], label='161 cow', color = 'green', marker = 'o')
plt.plot(cow171['Month after calving'], cow171['Diff'], label='at MIM = 7', color = 'pink', marker = 'o')
# plt.plot(cow181['Month after calving'], cow181['Diff'], label='181 cow', color = 'grey', marker = 'o')
# plt.plot(cow191['Month after calving'], cow191['Diff'], label='191 cow', color = 'c', marker = 'o')
plt.plot(cow1101['Month after calving'], cow1101['Diff'], label='at MIM = 10', color = 'orange', marker = 'o')
# plot all finally pick two

# always open cow
plt.plot(open1['Month after calving'], open1['Diff'], label='Parity 1 open cow', color = 'black', marker = 'o')

# Add labels and title
plt.xlabel('Month after calving')
plt.ylabel('''Q-value difference between 'keep' and 'cull'  ''')
plt.title('Value of cows who get pregnant at different month (Parity = 1)')
plt.legend()
# Set x-axis to use integer ticks
plt.xticks(df['mim'].unique())

plt.axhline(y=0, color='black', linestyle='dashed')

plt.show()
```
